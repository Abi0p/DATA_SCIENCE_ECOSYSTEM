```python
import pandas as pd
import numpy as np
data = pd.read_csv("census_income_data.csv")
```

```python
data.head()
```

```python
data.info()#check data types and missing values
```

```python
#check for missing values in the dataset
missing_values = data.isnull().sum()
missing_values
```

```python
#summary
data.describe()
```

```python
#Data cleaning
missing_percentage = (data.isnull().sum()/len(data))*100
missing_percentage = missing_percentage[missing_percentage>0].sort_values(ascending=False)
missing_percentage
```

```python
data_cleaned = data.drop(columns=['MIGMTR1','MIGMTR3','MIGMTR4','MIGSUN'])
data_cleaned = data_cleaned.dropna()
data_cleaned.info()
```

```python
print(data_cleaned.columns)
```

```python
#Highlight Insights Related to Demographic Influences on Income
import seaborn as sns
import matplotlib.pyplot as plt

# Age vs Income
plt.figure(figsize=(10, 6))
sns.countplot(x ='income_bracket', hue ='AAGE', data = data_cleaned)
plt.title('Age vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Age')
plt.show()

# Gender vs Income
plt.figure(figsize=(10, 6))
sns.countplot(x='income_bracket', hue='ASEX', data=data_cleaned)
plt.title('Gender vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Race vs Income
plt.figure(figsize=(10, 6))
sns.countplot(x='income_bracket', hue='ARACE', data=data_cleaned)
plt.title('Race vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Count')
plt.legend(title='Race')
plt.show()
```

```python
# Impact of Socioeconomic Factors on Income
plt.figure(figsize=(10, 6))
sns.countplot(y='AHGA', hue='income_bracket', data=data_cleaned, order=data_cleaned['AHGA'].value_counts().index)
plt.title('Education vs Income')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.legend(title='Income Bracket')
plt.show()
```

```python
# Employment Type vs Income
plt.figure(figsize=(10, 6))
sns.countplot(y='ACLSWKR', hue='income_bracket', data=data_cleaned, order=data_cleaned['ACLSWKR'].value_counts().index)
plt.title('Employment Type vs Income')
plt.xlabel('Count')
plt.ylabel('Employment Type')
plt.legend(title='Income Bracket')
plt.show()
```

```python
#Family and Household Dynamics and Their Correlation with Income
plt.figure(figsize=(10, 6))
sns.countplot(y='HHDFMX', hue='income_bracket', data=data_cleaned, order=data_cleaned['HHDFMX'].value_counts().index)
plt.title('Household Dynamics vs Income')
plt.xlabel('Count')
plt.ylabel('Household Status')
plt.legend(title='Income Bracket')
plt.show()
```

```python
# Number of Employers vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='income_bracket', y='NOEMP', data=data_cleaned)
plt.title('Number of Employers vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Number of Employers')
plt.show()
```

```python
# Region of Previous Residence vs Income
plt.figure(figsize=(10, 6))
sns.countplot(y='GRINREG', hue='income_bracket', data=data_cleaned, order=data_cleaned['GRINREG'].value_counts().index)
plt.title('Region of Previous Residence vs Income')
plt.xlabel('Count')
plt.ylabel('Region of Previous Residence')
plt.legend(title='Income Bracket')
plt.show()

# State of Previous Residence vs Income
plt.figure(figsize=(10, 6))
sns.countplot(y='GRINST', hue='income_bracket', data=data_cleaned, order=data_cleaned['GRINST'].value_counts().index)
plt.title('State of Previous Residence vs Income')
plt.xlabel('Count')
plt.ylabel('State of Previous Residence')
plt.legend(title='Income Bracket')
plt.show()

# Migration within region vs Income
plt.figure(figsize=(10, 6))
sns.countplot(x='MIGSAME', hue='income_bracket', data=data_cleaned)
plt.title('Migration within Region vs Income')
plt.xlabel('Same Residence as Last Year')
plt.ylabel('Count')
plt.legend(title='Income Bracket')
plt.show()
```

```python
# Capital Gains vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='income_bracket', y='CAPGAIN', data=data_cleaned)
plt.title('Capital Gains vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Capital Gains')
plt.show()

# Capital Losses vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='income_bracket', y='GAPLOSS', data=data_cleaned)
plt.title('Capital Losses vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Capital Losses')
plt.show()

# Dividend Income vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='income_bracket', y='DIVVAL', data=data_cleaned)
plt.title('Dividend Income vs Income')
plt.xlabel('Income Bracket')
plt.ylabel('Dividend Income')
plt.show()
```
